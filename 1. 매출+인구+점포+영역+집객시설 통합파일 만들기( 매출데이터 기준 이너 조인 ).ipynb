- 집객시설 null값 0으로 치환

```python
# null 값을 0으로 치환
facility = facility.fillna(0)

# 확인 (선택)
print(facility.head())
```

- 집객시설 2020 4분기 데이터로 2019 1분기 ~ 2020 3분기 데이터 만들어주기

```python
# 1. 복사할 기준 데이터 (2020년 4분기)
reference_data = facility[facility['기준_년분기_코드'] == 20204]

# 2. 복사할 대상 분기 목록 (2019년 1분기 ~ 2020년 3분기)
missing_quarters = [20191, 20192, 20193, 20194, 20201, 20202, 20203]

# 3. 기준 데이터를 복사하여 누락된 분기에 대한 데이터 생성
new_rows = []
for quarter in missing_quarters:
    temp = reference_data.copy()
    temp['기준_년분기_코드'] = quarter
    new_rows.append(temp)

# 4. 새로운 데이터(2019년 1분기~202년 3분기)를 기존 데이터프레임에 추가
new_data = pd.concat(new_rows, ignore_index=True)
facility_updated = pd.concat([facility, new_data], ignore_index=True)

# 결과 확인
print("결측 분기를 추가한 데이터프레임:")
facility_updated.sort_values(by=['기준_년분기_코드'], inplace=True)
facility_updated.reset_index(drop=True, inplace=True)
facility_updated.head()

# 저장 (필요한 경우)
# facility_updated.to_csv('집객시설_결측분기채움.csv', index=False, encoding='utf-8-sig')

```

```python
facility_not_null = facility_updated
```

```python
# 매출 + 인구 데이터

# revenue 기준으로 population 데이터 병합
merged_data = revenue.merge(
    population,  # population 전체 사용
    on=['기준_년분기_코드', '상권_코드_명'],  # 조인키 기준으로 병합
    how='inner',
    suffixes=('', '_remove')  # 중복 컬럼 처리: population에서 온 중복 컬럼에 '_remove' 접미사 추가
)

# 중복 컬럼 제거 (_remove 접미사가 붙은 컬럼 삭제)
merged_data = merged_data.loc[:, ~merged_data.columns.str.endswith('_remove')]

# 결과 확인
print("병합된 데이터프레임 정보:")
print(merged_data.info())
print(merged_data.shape)
print(revenue.shape)

# 저장 (필요한 경우)
# merged_data.to_csv('merged_revenue_population.csv', index=False, encoding='utf-8-sig')

```

```python
# (매출+인구) + 집객시설데이터

# merged_data(매출+인구) 기준으로 facility 데이터 병합
merged_data1 = merged_data.merge(
    facility_not_null,  # facility_not_null 전체 사용
    on=['기준_년분기_코드', '상권_코드_명'],  # 조인키 기준으로 병합
    how='inner',
    suffixes=('', '_remove')  # 중복 컬럼 처리: facility_not_null에서 온 중복 컬럼에 '_remove' 접미사 추가
)

# 중복 컬럼 제거 (_remove 접미사가 붙은 컬럼 삭제)
merged_data1 = merged_data1.loc[:, ~merged_data1.columns.str.endswith('_remove')]

# 결과 확인
print("병합된 데이터프레임 정보:")
print(merged_data1.info())
print(merged_data1.shape)
print(merged_data.shape)
```

```python
# (매출+인구 + 집객시설) + 점포데이터

# merged_data1(매출+인구+집객시설) 기준으로 store 데이터 병합
merged_data2 = merged_data1.merge(
    store,  # store 전체 사용
    on=['기준_년분기_코드', '상권_코드_명', '서비스_업종_코드_명'],  # 조인키 기준으로 병합
    how='inner',
    suffixes=('', '_remove')  # 중복 컬럼 처리: store에서 온 중복 컬럼에 '_remove' 접미사 추가
)

# 중복 컬럼 제거 (_remove 접미사가 붙은 컬럼 삭제)
merged_data2 = merged_data2.loc[:, ~merged_data2.columns.str.endswith('_remove')]

# 결과 확인
print("병합된 데이터프레임 정보:")
print(merged_data2.info())
print(merged_data2.shape) # 합친데이터
print(merged_data1.shape) # 이전데이터 행수 비교(얼마나 줄어들었는지)
```

```python
# (매출 + 인구 + 집객시설 + 점포데이터) + 영역

# merged_data1(매출+인구+집객시설) 기준으로 area데이터 병합
merged_data3 = merged_data2.merge(
    area,  # area 전체 사용
    on= ['상권_코드_명'],  # 조인키 기준으로 병합
    how='inner',
    suffixes=('', '_remove')  # 중복 컬럼 처리: area에서 온 중복 컬럼에 '_remove' 접미사 추가
)

# 중복 컬럼 제거 (_remove 접미사가 붙은 컬럼 삭제)
merged_data3 = merged_data3.loc[:, ~merged_data3.columns.str.endswith('_remove')]

# 결과 확인
print("병합된 데이터프레임 정보:")
print(merged_data3.info())
print(merged_data3.shape)
print(merged_data2.shape)
```
